```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
#gr√°fico logaritmo

x = np.logspace(-1, 1, 40)
y1 = x**2.0
y2 = x**1.5
plt.loglog(x, y1, "bo-", linewidth=2, markersize=12, label='First')
plt.loglog(x, y2, "gs-", linewidth=2, markersize=12, label='Second')
plt.xlabel('X')
plt.ylabel('Y')
#plt.axis([-0.5, 10.5, -5, 105])
plt.legend(loc="upper left")
plt.show()
```

```python
#histograma

x = np.random.normal(size=1000)
plt.hist(x, density=True,  bins=10)
plt.show()
```

```python
#histogramas normalizados e cumulativos
x = np.random.gamma(2, 3, 100000)

plt.hist(x, bins=30, density=True)
plt.title('Grafico 1 - normalizado')
plt.show()
plt.hist(x, bins=30, cumulative=True)
plt.title('Grafico 2 - cumulative')
plt.show()
plt.hist(x, bins=30, density=True, cumulative=True, histtype="step")
plt.title('Grafico 3 - normalizado and cumulative')
plt.show()
```

```python
#usando subplot
plt.subplot(221)
plt.hist(x, bins=30)
plt.subplot(222)
plt.hist(x, bins=30, density=True)
plt.subplot(223)
plt.hist(x, bins=30, cumulative=True)
plt.subplot(224)
plt.hist(x, bins=30, density=True, cumulative=True, histtype="step")
plt.show()
```
